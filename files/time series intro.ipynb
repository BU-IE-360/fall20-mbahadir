# Time-series examples
## Electricity consumption
[Transparency Platform](https://seffaflik.epias.com.tr/transparency/) provides several time-series data related to the electricity market. Turkey's electricity consumption can be reached from the [real time consumption link](https://seffaflik.epias.com.tr/transparency/tuketim/gerceklesen-tuketim/gercek-zamanli-tuketim.xhtml). This page allows for filtering options based on the date. Note that recent consumption values (up to 3 days) are sometimes updated by the authorities. We will consider the consumption data from 2019-01-01 to 2020-11-09. The data is downloaded as csv after filtering the dates. The examples will mainly benefit from data.table package for data manipulation, lubridate package for time-based manipulations and ggplot2 package for visualization purposes.

```r
library(data.table,quietly = TRUE)
library(ggplot2,quietly = TRUE)
library(lubridate,quietly = TRUE)

data_path='/mnt/11a4ca17-594d-4aab-b29d-9e8c3dcbbce8/Courses/IE360/Fall20/codes/jupyter/data/GercekZamanliTuketim-01012019-09112020.csv'

consumption=fread(data_path,encoding='UTF-8')
setnames(consumption,names(consumption),c('Date','Hour','Consumption'))
head(consumption,10)
#manipulating date values
consumption[,Date:=as.Date(Date,'%d.%m.%Y')]
#adding timestamp
consumption[,timestamp:=ymd_hm(paste(Date,Hour))]
head(consumption,10)
```

```r
#manipulating consumption values
consumption[,Consumption:=gsub('\\.','',Consumption)]
consumption[,Consumption:=gsub('\\,','.',Consumption)]
consumption[,Consumption:=as.numeric(Consumption)]

head(consumption,10)
```

```r
ggplot(data=consumption, aes(x=timestamp, y=Consumption)) + geom_line()
```

```r
ggplot(data=tail(consumption,168), aes(x=timestamp, y=Consumption)) + geom_line()
```
